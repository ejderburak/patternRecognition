Collecting Data for the Data Set Using OpenCV

Purpose of this code is taking pictures with webcam and saving it to a destined file.

```python
import cv2
import time
import os
from datetime import datetime

# Setting for each hand gesture
GESTURE_NAME = "stop"        # approve, ok, peace...
PERSON_TAG = "burak_stop"    # zeynep, burak (optinial)
PHOTOS_TO_TAKE = 50          # number of the photos to take
INTERVAL_SEC = 1.0           # time between each shot

SAVE_DIR = f"photos/{GESTURE_NAME}"
os.makedirs(SAVE_DIR, exist_ok=True)

# put camera id as parameter
cap = cv2.VideoCapture(0)

print("Waiting for webcam to open...")
time.sleep(2)
for _ in range(10):
    cap.read()

print(f"Starting: {PHOTOS_TO_TAKE} pics | gesture={GESTURE_NAME} | file={SAVE_DIR}")

count = 0
while count < PHOTOS_TO_TAKE:
    ret, frame = cap.read()
    if not ret:
        print("Couldnot take a photo, trying again")
        continue

    ts = datetime.now().strftime("%Y%m%d_%H%M%S_%f")
    filename = f"{GESTURE_NAME}_{PERSON_TAG}_{ts}.jpg"
    path = os.path.join(SAVE_DIR, filename)

    cv2.imwrite(path, frame)
    count += 1
    print(f"{count}/{PHOTOS_TO_TAKE} saved to: {filename}")

    time.sleep(INTERVAL_SEC)

cap.release()
print(f"Finished! {count} picture have been saved -> {SAVE_DIR}")
```
